## **ðŸŽ¯ Objetivo**
1. El jugador intenta adivinar un nÃºmero aleatorio entre 1 y 100.
2. Se registra cada intento en una lista.
3. Al finalizar, se realiza un anÃ¡lisis de los intentos:
   - Cantidad total de intentos.
   - Promedio de intentos.
   - NÃºmero mÃ¡ximo y mÃ­nimo de intentos.
   - Histograma de distribuciÃ³n de intentos.  

---

## **ðŸ“Œ CÃ³digo del Juego**
```python
import random
import pandas as pd
import matplotlib.pyplot as plt

def juego_adivinanza():
    # NÃºmero aleatorio entre 1 y 100
    numero_secreto = random.randint(1, 100)
    
    intentos = []
    intento = None

    print("ðŸŽ® Bienvenido al Juego de Adivinanza ðŸŽ®")
    print("Adivina un nÃºmero entre 1 y 100")

    while intento != numero_secreto:
        try:
            intento = int(input("Introduce tu nÃºmero: "))
            intentos.append(intento)

            if intento < numero_secreto:
                print("ðŸ”¼ Muy bajo, intenta de nuevo.")
            elif intento > numero_secreto:
                print("ðŸ”½ Muy alto, intenta de nuevo.")
            else:
                print("ðŸŽ‰ Â¡Felicidades! Adivinaste el nÃºmero.")
        except ValueError:
            print("âŒ Entrada invÃ¡lida, introduce un nÃºmero.")

    # AnÃ¡lisis de datos
    df = pd.DataFrame(intentos, columns=["Intentos"])
    
    print("\nðŸ“Š AnÃ¡lisis de Datos ðŸ“Š")
    print(f"ðŸ”¹ Total de intentos: {len(intentos)}")
    print(f"ðŸ”¹ Promedio de intentos: {df['Intentos'].mean():.2f}")
    print(f"ðŸ”¹ Intento mÃ¡s alto: {df['Intentos'].max()}")
    print(f"ðŸ”¹ Intento mÃ¡s bajo: {df['Intentos'].min()}")

    # VisualizaciÃ³n del histograma de intentos
    plt.hist(df["Intentos"], bins=10, color='blue', alpha=0.7, edgecolor='black')
    plt.xlabel("NÃºmeros intentados")
    plt.ylabel("Frecuencia")
    plt.title("Histograma de Intentos")
    plt.show()

# Ejecutar el juego
juego_adivinanza()
```

---

## **ðŸ“Œ ExplicaciÃ³n del CÃ³digo**
1. **Se genera un nÃºmero aleatorio** entre 1 y 100.  
2. **El jugador introduce nÃºmeros** hasta adivinar el correcto.  
3. **Se almacenan los intentos** en una lista.  
4. **Se analizan los intentos con pandas**: total, promedio, mÃ¡ximo y mÃ­nimo.  
5. **Se visualizan los intentos con un histograma en matplotlib**.

## **ðŸ“Œ PrÃ¡ctica de clase**
### **ðŸ•’ AplicaciÃ³n 1: Fundamentos de Pandas y Carga de Datos**
#### ðŸ“ Ejercicio 1: Cargar y Explorar Datos  
- **Objetivo:** Cargar un dataset en Pandas y realizar exploraciÃ³n inicial.  
- **Dataset:** Titanic (o cualquier otro dataset pÃºblico)  

```python
import pandas as pd  

# Cargar dataset desde una URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"  
df = pd.read_csv(url)  

# Mostrar informaciÃ³n del dataset
print(df.head())  # Primeras 5 filas
print(df.info())  # Tipos de datos
print(df.describe())  # EstadÃ­sticas bÃ¡sicas
```
**ðŸ“Œ Preguntas:**  
1. Â¿CuÃ¡ntas filas y columnas tiene el dataset?  
2. Â¿CuÃ¡les son los tipos de datos de cada columna?  
3. Â¿Hay valores nulos?  

---

### **ðŸ•’ AplicaciÃ³n 2: Limpieza y TransformaciÃ³n de Datos**
#### ðŸ“ Ejercicio 2: Manejo de Valores Nulos  
- **Objetivo:** Identificar y manejar valores nulos.  

```python
# Contar valores nulos por columna
print(df.isnull().sum())

# Rellenar la columna "Age" con la media
df["Age"].fillna(df["Age"].mean(), inplace=True)

# Rellenar la columna "Embarked" con el valor mÃ¡s frecuente
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Verificar nuevamente valores nulos
print(df.isnull().sum())
```
**ðŸ“Œ Preguntas:**  
1. Â¿CuÃ¡ntos valores nulos tenÃ­a la columna "Age" antes de la limpieza?  
2. Â¿CÃ³mo afecta la media a los datos originales?  

#### ðŸ“ Ejercicio 3: CreaciÃ³n de Nuevas Columnas  
- **Objetivo:** Crear nuevas columnas a partir de las existentes.  

```python
# Crear una columna que categorice la edad en niÃ±os, jÃ³venes y adultos
df["Age_Group"] = pd.cut(df["Age"], bins=[0, 12, 18, 99], labels=["NiÃ±o", "Joven", "Adulto"])
print(df[["Age", "Age_Group"]].head(10))
```
**ðŸ“Œ Pregunta:** Â¿CÃ³mo cambiarÃ­an los resultados si modificamos los rangos?

### **ðŸ•’ AplicaciÃ³n 3: AnÃ¡lisis Exploratorio y VisualizaciÃ³n de Datos**
#### ðŸ“ Ejercicio 4: DistribuciÃ³n de la Edad  
- **Objetivo:** Graficar la distribuciÃ³n de edad con un histograma.  

```python
import matplotlib.pyplot as plt  

plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")  
plt.xlabel("Edad")  
plt.ylabel("Frecuencia")  
plt.title("DistribuciÃ³n de Edad en el Titanic")  
plt.show()
```
**ðŸ“Œ Pregunta:** Â¿CuÃ¡l es el rango de edad mÃ¡s comÃºn en los pasajeros?  

#### ðŸ“ Ejercicio 5: Supervivencia por GÃ©nero  
- **Objetivo:** Comparar la tasa de supervivencia entre hombres y mujeres.  

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar", color=["blue", "pink"])
plt.xlabel("GÃ©nero")
plt.ylabel("Tasa de Supervivencia")
plt.title("Supervivencia en el Titanic por GÃ©nero")
plt.show()
```
**ðŸ“Œ Pregunta:** Â¿QuÃ© gÃ©nero tuvo mayor tasa de supervivencia? Â¿Por quÃ©?  

---

### **ðŸ•’ AplicaciÃ³n 4: Correlaciones y AnÃ¡lisis Avanzado**
#### ðŸ“ Ejercicio 6: RelaciÃ³n entre Clase y Supervivencia  
- **Objetivo:** Analizar cÃ³mo influyÃ³ la clase del pasajero en la supervivencia.  

```python
df.groupby("Pclass")["Survived"].mean().plot(kind="bar", color=["blue", "orange", "red"])
plt.xlabel("Clase")
plt.ylabel("Tasa de Supervivencia")
plt.title("Supervivencia por Clase en el Titanic")
plt.show()
```
**ðŸ“Œ Pregunta:** Â¿Los pasajeros de primera clase tuvieron mayor probabilidad de sobrevivir?  

#### ðŸ“ Ejercicio 7: Matriz de CorrelaciÃ³n  
- **Objetivo:** Encontrar correlaciones entre variables.  

```python
import seaborn as sns  

correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de CorrelaciÃ³n en el Titanic")
plt.show()
```
**ðŸ“Œ Pregunta:** Â¿QuÃ© variables tienen mayor relaciÃ³n con la supervivencia?
