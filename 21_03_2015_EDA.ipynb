### **2. Carga y Exploraci贸n de Datos (30 min)**  
 **Ejercicio:** Cargar y visualizar la estructura del dataset.  
```python
import pandas as pd  

# Cargar datos  
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")  

# Mostrar primeras filas  
print(df.head())  

# Informaci贸n general  
print(df.info())  

# Descripci贸n estad铆stica  
print(df.describe())  
```

### **3. Limpieza y Tratamiento de Datos (30 min)**  
 **Ejercicio:** Manejo de valores nulos y transformaci贸n de variables.  
```python
# Verificar valores nulos  
print(df.isnull().sum())  

# Rellenar valores nulos en 'Age' con la mediana  
df['Age'].fillna(df['Age'].median(), inplace=True)  

# Rellenar 'Embarked' con el valor m谩s com煤n  
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  

# Eliminar columna innecesaria  
df.drop(columns=['Cabin'], inplace=True)  
```
 **Discusi贸n:**  
- 驴C贸mo afectan los valores nulos a nuestro an谩lisis?

### **4. An谩lisis de Variables y Visualizaci贸n (45 min)**  
 **Ejercicio:** Distribuci贸n de variables y correlaciones.  

 **Visualizar la distribuci贸n de edades:**  
```python
import seaborn as sns  
import matplotlib.pyplot as plt  

sns.histplot(df['Age'], bins=30, kde=True)  
plt.title("Distribuci贸n de edades")  
plt.show()  
```

 **Supervivencia por g茅nero:**  
```python
sns.countplot(x="Sex", hue="Survived", data=df)  
plt.title("Supervivencia por g茅nero")  
plt.show()  
```

 **Correlaci贸n entre variables:**  
```python
import numpy as np  

corr_matrix = df.corr()  
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)  
plt.title("Matriz de correlaci贸n")  
plt.show()  
```
 **Discusi贸n:**  
- 驴Qu茅 insights obtenemos de los gr谩ficos?

### **5. An谩lisis Avanzado: Relaci贸n de Variables con la Supervivencia (45 min)**  
 **Ejercicio:** Identificar patrones entre clase, edad y supervivencia.  

 **Supervivencia por clase y edad:**  
```python
sns.boxplot(x="Pclass", y="Age", hue="Survived", data=df)  
plt.title("Edad y supervivencia por clase")  
plt.show()  
```

 **Relaci贸n entre tarifa del boleto y supervivencia:**  
```python
sns.histplot(df[df["Survived"]==1]["Fare"], bins=30, kde=True, color="green", label="Sobrevivientes")  
sns.histplot(df[df["Survived"]==0]["Fare"], bins=30, kde=True, color="red", label="No sobrevivientes")  
plt.legend()  
plt.title("Distribuci贸n de tarifas por supervivencia")  
plt.show()  
```
